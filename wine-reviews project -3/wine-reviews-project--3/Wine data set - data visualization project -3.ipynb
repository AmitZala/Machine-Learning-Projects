# Wine data set - data visualization project -3 

This data set take from kaggle - https://www.kaggle.com/zynicide/wine-reviews

```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing dataset 
wine = pd.read_csv('winemag-data_first150k.csv', index_col=0)
wine.head()
```

Using Matplotlib for data visulization


1. Histogram

```python
# Create figure and axis
fig, ax = plt.subplots()

# Plot histogram
ax.hist(wine['points'])

# set title and labels
ax.set_title('Wine Review Scores')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

2. Bar chart


```python
# Creat a figure and axis
fig, ax = plt.subplots()

# count the occurrence of each class
data = wine['points'].value_counts()

# Get X and Y data 
pt = data.index
fre = data.values

# Create bar chart
ax.bar(pt, fre)

# Set title and lables
ax.set_title('Wine Review Scores')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

3. Histogram

4. Bar Charts

```python
wine['points'].value_counts().sort_index().plot.bar()
```

```python
# Horizontal bar-chart
wine['points'].value_counts().sort_index().plot.barh() 
```

```python
# Grouped the data by country and then took the mean of the wine prices
wine.groupby("country").price.mean().sort_values(ascending=False)[:5].plot.bar()
```

Data visualization using seaborn

5. Histogram

```python
#Importing Seaborn Libraries
import seaborn as sns
```

```python
sns.distplot(wine['points'], bins=10, kde=False)
```

```python
#Histogram with gaussian kernel density estimate
sns.distplot(wine['points'], bins=10, kde=True)
```

6. Bar chart

```python
sns.countplot(wine['points'])
```

7. Box plots

```python
df = wine[(wine['points']>=95) & (wine['price']<1000)]
sns.boxplot('points', 'price', data=df)
```

```python
sns.heatmap(wine.corr(), annot=True)
```
